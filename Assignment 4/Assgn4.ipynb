```python

import pandas as pd
#data in sf and daly city between 01/01/21 and 01/01/24
crashes_sf = pd.read_csv('Crashes_SF.csv')
crashes_dc = pd.read_csv('Crashes_DC.csv')
parties_sf = pd.read_csv('Parties_SF.csv')
parties_dc = pd.read_csv('Parties_DC.csv')
```

```python
print(parties_sf['INATTENTION'].unique())
```

```python
parties_sf_filter = parties_sf[(parties_sf['INATTENTION'] != '-') & (parties_sf['INATTENTION'].notna())]
parties_dc_filter = parties_dc[(parties_dc['INATTENTION'] != '-') & (parties_dc['INATTENTION'].notna())]
```

```python
parties_sf_filter
```

```python
crashes_sf_severity = crashes_sf[crashes_sf['COLLISION_SEVERITY'].isin([1,2,3])]
crashes_dc_severity = crashes_dc[crashes_dc['COLLISION_SEVERITY'].isin([1,2,3])]
crashes_sf_severity
```

```python
crashes_sf_filtered = crashes_sf_severity[crashes_sf_severity['PCF_VIOL_CATEGORY'].isin(['01','24'])]
crashes_dc_filtered = crashes_dc_severity[crashes_dc_severity['PCF_VIOL_CATEGORY'].isin(['01','24'])]
crashes_dc_filtered
```

```python
final_df_sf = pd.merge(crashes_sf_filtered, parties_sf_filter, on='CASE_ID', how='left')
final_df_dc = pd.merge(crashes_dc_filtered, parties_dc_filter, on='CASE_ID', how='left')
```

```python
final_df_dc
```

```python
crash_df = pd.concat([final_df_sf, final_df_dc], axis=0)
crash_df = crash_df.dropna(subset=['POINT_X', 'POINT_Y'])
crash_df['POINT_Y'].isna().sum()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from scipy.stats import gaussian_kde
import contextily as ctx

def apply_kde_no_autocorrelation(crash_df, bandwidth=0.05):
    """Applies Kernel Density Estimation (KDE) to crash locations without considering autocorrelation."""
    crash_points = crash_df[['POINT_X', 'POINT_Y']].values
    weights = np.ones(len(crash_points))  # Each crash has equal weight
    
    # Perform KDE
    kde = gaussian_kde(crash_points.T, weights=weights, bw_method=bandwidth)
    
    # Create a meshgrid for visualization
    buffer = 0.005
    x_min, x_max = crash_points[:, 0].min() - buffer, crash_points[:, 0].max() + buffer
    y_min, y_max = crash_points[:, 1].min() - buffer, crash_points[:, 1].max() + buffer
    x_grid, y_grid = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    grid_coords = np.vstack([x_grid.ravel(), y_grid.ravel()])
    density = kde(grid_coords).reshape(x_grid.shape)
    
    # Plot the KDE heatmap
    plt.figure(figsize=(10, 8))
    #plt.imshow(density, origin='lower', extent=[x_min, x_max, y_min, y_max], 
     #      cmap='Reds', alpha=0.5, vmin=0, vmax=np.percentile(density, 95)) 
    plt.imshow(density, origin='lower', extent=[x_min, x_max, y_min, y_max], cmap='Reds', alpha=0.75)

    plt.colorbar(label='Crash Density')

    plt.scatter(crash_points[:, 0], crash_points[:, 1], c='blue', s=10, alpha=0.5, label='Crash Locations')
    ctx.add_basemap(plt.gca(), crs='EPSG:4326', source=ctx.providers.CartoDB.PositronNoLabels, zorder=-1)
    plt.title("Kernel Density Estimation of Crashes (No Autocorrelation)")
    plt.legend()
    plt.show()






```

```python
# Jupyter Notebook Execution
apply_kde_no_autocorrelation(crash_df)
```

```python
def apply_dbscan(crash_df, eps=0.005, min_samples=3):
    """Applies DBSCAN clustering to crash locations."""
    crash_points = crash_df[['POINT_X', 'POINT_Y']].values
    
    # Apply DBSCAN
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(crash_points)
    
    # Plot results
    plt.figure(figsize=(8, 8))
    unique_labels = set(labels)
    colors = plt.cm.jet(np.linspace(0, 1, len(unique_labels)))
    
    for label, color in zip(unique_labels, colors):
        if label == -1:
            color = 'black'  # Noise points
        cluster_points = crash_points[labels == label]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=[color], s=20, label=f'Cluster {label}')
    
    plt.title("DBSCAN Clustering of Crash Locations")
    plt.legend()
    ctx.add_basemap(plt.gca() , crs='EPSG:4326', source=ctx.providers.CartoDB.PositronNoLabels)
    plt.show()
```

```python
apply_dbscan(crash_df)
```
